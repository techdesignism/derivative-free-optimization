```python

from numpy import *
import matplotlib.pyplot as plt
from obj_funcs_lib import *


import numpy as np
np.set_printoptions(precision=2)
import nevergrad as ng
from obj_funcs_lib import *


import time
```

```python
with open("valid_opt_list_rosenbrock.txt", "r") as f:
    filtered_optimizers = [line.strip() for line in f if line.strip()]


all_optimizers = list(ng.optimizers.registry.keys())

filtered_all_optimizers = [name for name in all_optimizers if name in filtered_optimizers]

print(len(filtered_all_optimizers))
filtered_all_optimizers
```

```python
dim = 3
start_point = [200] * dim

current_obj = rosenbrock
i = 0
for opt_name in filtered_all_optimizers:
    print(i)

    try:
        start_time = time.time()
        optimizer_cls = ng.optimizers.registry[opt_name]
        optimizer = optimizer_cls(parametrization=ng.p.Array(init=start_point), budget=5000)
        recommendation = optimizer.minimize(lambda x: current_obj(x))
        best_value = current_obj(recommendation.value)
        elapsed = time.time() - start_time
        print(f"Best value for {opt_name}: {best_value}")
        print(f"Time taken for {opt_name}: {elapsed:.2f} seconds")
        print('='*20)
    except Exception as e:
        print(e)
        pass
    i += 1
```
