```python
# Use nevergrad Powell optimizer to optimize the sphere function from obj_funcs_lib, with a start point

# Problem definition
dim = 3  # dimension of the problem
start_point = [200] * dim

# Do NOT set bounds
instrum = ng.p.Array(init=start_point)
optimizer = ng.optimizers.Powell(parametrization=instrum, budget=10000)

def evaluate(x):
    # x is usually either an Instrumentation or a numpy array depending on the optimizer
    # Here, for ng.p.Array, x is np.ndarray directly
    return ackley(x)

recommendation = optimizer.minimize(evaluate)
print("Best value found:", sphere(recommendation.value))
print("Best parameters:", recommendation.value)
```
